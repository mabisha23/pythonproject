```python
import numpy as np
import pandas as pd
df = pd.read_csv("cl.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder

# Select categorical columns
cat_cols = df.select_dtypes(include=['object']).columns

# Encode categorical columns, ignoring null values
for col in cat_cols:
    le = LabelEncoder()  # Create a new LabelEncoder for each column
    non_null_values = df[col][df[col].notnull()]  # Exclude null values for fitting
    le.fit(non_null_values)  # Fit only on non-null values

    # Transform the column, leaving nulls unchanged
    df[col] = df[col].apply(lambda x: le.transform([x])[0] if pd.notnull(x) else x)

# Display the updated DataFrame
df
```

```python
df.isnull().sum()
```

```python
iris_data = df
# Assuming 'heart_data' is your DataFrame
# Separate numeric and non-numeric columns
numeric_columns = iris_data.select_dtypes(include=['number']).columns
non_numeric_columns = iris_data.select_dtypes(exclude=['number']).columns

# Fill missing values in numeric columns with mean or median
iris_data[numeric_columns] = iris_data[numeric_columns].fillna(iris_data[numeric_columns].mean())
# Alternatively, to use the median:
# heart_data[numeric_columns] = heart_data[numeric_columns].fillna(heart_data[numeric_columns].median())

# Fill missing values in non-numeric columns with mode
for column in non_numeric_columns:
    iris_data[column].fillna(iris_data[column].mode()[0], inplace=True)
```

```python
iris_data.isnull().sum()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
dt = DecisionTreeClassifier()
x = df.drop(columns=['ContactLens'])
y = df['ContactLens']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=42)
dt.fit(x_train,y_train)

```

```python
y_pred = dt.predict(x_test)
accuracy_score(y_test,y_pred)
```
